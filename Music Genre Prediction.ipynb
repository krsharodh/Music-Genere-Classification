# Music Genre Prediciton

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.dummy import DummyClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.neural_network import MLPClassifier

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit

from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import MultiLabelBinarizer


class Music_Genre_Prediction:

    def __init__(self):
        self.learning_rate = 0.01
        self.mlp_max_iter = 1000
        self.hidden_layer_sizes = (1000, 5)

    def train(self, x_train, y_train, mnb):

        if mnb:
            # MUltinomial Naive Bayes
            self.MNB = MultinomialNB()
            self.MNB.fit(x_train, y_train)
        else:
            # Gaussian Naive Bayes
            self.GNB = GaussianNB()
            self.GNB.fit(x_train, y_train)

        # Logistic Regression
        self.LR = make_pipeline(
            StandardScaler(), LogisticRegression(max_iter=2000))
        self.LR.fit(x_train, y_train)

        # Multi Layer Perceptron
        model = MLPClassifier(hidden_layer_sizes=(
            1000, 5), solver='sgd', learning_rate_init=0.01, max_iter=1000)
        self.MLP = make_pipeline(StandardScaler(), model)
        self.MLP.fit(x_train, y_train)

        # Dummy CLassifier
        self.ZERO_R = DummyClassifier(strategy='most_frequent')
        self.ZERO_R.fit(x_train, y_train)

    def predict(self, x_test, mnb):

        if mnb:
            # Multinomial Naive Bayes
            self.MNB_prediciton = self.MNB.predict(x_test)
        else:
            # Gaussian Naive Bayes
            self.GNB_prediciton = self.GNB.predict(x_test)

        # Logistic Regression
        self.LR_prediction = self.LR.predict(x_test)

        # Multi Layer Perceptron
        self.MLP_prediciton = self.MLP.predict(x_test)

        # Zero R
        self.ZERO_R_prediciton = self.ZERO_R.predict(x_test)

    def evaluate(self, model, prediction, x_test, y_test):
        print("Accuracy: ", model.score(x_test, y_test))

        # Precision
        precision = precision_score(y_test, prediction, average='macro')
        print("Precision:", precision)

        # Recall
        recall = recall_score(y_test, prediction, average='macro')
        print("Recall:", recall)

        # F-Score
        f1 = f1_score(y_test, prediction, average='macro')
        print("F Score:", f1)

    def train_predict_evaluate(self, x_train, y_train, x_test, y_test, mnb):
        # Training
        self.train(x_train, y_train, mnb)

        # Prediction
        self.predict(x_test, mnb)

        # Evaluation
        print("\nZero R")
        self.evaluate(self.ZERO_R, self.ZERO_R_prediciton,
                      x_test, y_test)
        if mnb:
            # Multinomial Naive Bayes
            print("\nMultinomial Naive Bayes:")
            self.evaluate(self.MNB, self.MNB_prediciton,
                          x_test, y_test)
        else:
            print("\nGaussian Naive Bayes:")
            self.evaluate(self.GNB, self.GNB_prediciton,
                          x_test, y_test)
        print("\nLogistic Regression")
        self.evaluate(self.LR, self.LR_prediction,
                      x_test, y_test)
        print("\nMulti Layer Perceptron")
        self.evaluate(self.MLP, self.MLP_prediciton,
                      x_test, y_test)

    def import_data(self):
        # Import Training Features
        self.train_features = pd.read_csv("train_features.csv")
        self.train_labels = pd.read_csv("train_labels.csv")

        # Import Validation Features
        self.valid_features = pd.read_csv("valid_features.csv")
        self.valid_labels = pd.read_csv("valid_labels.csv")

        # Import Testing Features
        self.test_features = pd.read_csv("test_features.csv")

    def split_dataset(self):
        # Training Data
        self.train_data = pd.concat(
            [self.train_features, self.train_labels], axis=1)
        self.train_data = self.train_data.drop(columns="trackID")

        self.train_data_features = self.train_data.drop(columns="genre")
        self.train_data_labels = self.train_data["genre"]

        # Validation Data
        self.valid_data = pd.concat(
            [self.valid_features, self.valid_labels], axis=1)
        self.valid_data = self.valid_data.drop(columns="trackID")

        self.valid_data_features = self.valid_data.drop(columns="genre")
        self.valid_data_labels = self.valid_data["genre"]

        # Test Data
        self.test_data_features = self.test_features.drop(columns="trackID")

    def extract_features(self):
        # Training Features
        self.train_data_audio_features = pd.DataFrame(
            self.train_data_features.iloc[:, 8:156])
        self.train_data_text_features = pd.DataFrame(
            self.train_data_features.tags)
        self.train_data_metadata_features = pd.DataFrame(self.train_data_features
                                                         .filter(items=["loudness",
                                                                        "tempo",
                                                                        "key",
                                                                        "mode",
                                                                        "duration",
                                                                        "time_signature"]))
        self.train_data_title_features = pd.DataFrame(
            self.train_data_features.title)

        # Validation Features
        self.valid_data_audio_features = pd.DataFrame(
            self.valid_data_features.iloc[:, 8:156])
        self.valid_data_text_features = pd.DataFrame(
            self.valid_data_features.tags)
        self.valid_data_metadata_features = pd.DataFrame(self.valid_data_features
                                                         .filter(items=["loudness",
                                                                        "tempo",
                                                                        "key",
                                                                        "mode",
                                                                        "duration",
                                                                        "time_signature"]))
        self.valid_data_title_features = pd.DataFrame(
            self.valid_data_features.title)

        # Testing Features
        self.test_data_audio_features = pd.DataFrame(
            self.test_data_features.iloc[:, 8:156])
        self.test_data_text_features = pd.DataFrame(
            self.test_data_features.tags)
        self.test_data_metadata_features = pd.DataFrame(self.test_data_features
                                                        .filter(items=["loudness",
                                                                       "tempo",
                                                                       "key",
                                                                       "mode",
                                                                       "duration",
                                                                       "time_signature"]))
        self.test_data_title_features = pd.DataFrame(
            self.valid_data_features.title)

    def pre_process_tags(self):

        mlb = MultiLabelBinarizer()

        train_tag_convert = mlb.fit_transform(
            self.train_data_text_features['tags'].apply(lambda x: set(x.split(','))))
        train_tags_tfid = pd.DataFrame(train_tag_convert, columns=mlb.classes_)

        valid_tag_convert = mlb.transform(
            self.valid_data_text_features['tags'].apply(lambda x: set(x.split(','))))
        valid_tags_tfid = pd.DataFrame(valid_tag_convert, columns=mlb.classes_)

        SKB = SelectKBest(k=300)
        self.kbest_Tags = SKB.fit_transform(
            train_tags_tfid, self.train_data_labels)
        self.kbest_validationTags = SKB.transform(valid_tags_tfid)

    def pre_process_title(self):

        mlb = MultiLabelBinarizer()

        train_title_convert = mlb.fit_transform(
            self.train_data_title_features['title'].apply(lambda x: set(x.split(' '))))
        train_title_tfid = pd.DataFrame(
            train_title_convert, columns=mlb.classes_)

        valid_title_convert = mlb.transform(
            self.valid_data_title_features['title'].apply(lambda x: set(x.split(' '))))
        valid_title_tfid = pd.DataFrame(
            valid_title_convert, columns=mlb.classes_)

        SKB = SelectKBest(k=300)
        self.kbest_title = SKB.fit_transform(
            train_title_tfid, self.train_data_labels)
        self.kbest_validationTitle = SKB.transform(valid_title_tfid)

    def draw_curves(self, estimator, X, y, ylim, cv,
                    n_jobs, train_sizes=np.linspace(.1, 1.0, 5)):

        plt.figure()
        plt.xlabel("Training Size")
        plt.ylabel("Accuracy")
        plt.ylim(*ylim)

        train_sizes, train_scores, test_scores, fit_times, _ = \
            learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                           train_sizes=train_sizes,
                           return_times=True)
        train_scores_mean = np.mean(train_scores, axis=1)

        test_scores_mean = np.mean(test_scores, axis=1)
        test_scores_std = np.std(test_scores, axis=1)

        plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
        plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
        plt.legend(loc="best")

        return plt

    def export_test_labels(self):
        df = pd.DataFrame(
            {"trackID": self.test_features['trackID'], "genre": self.LR.predict(self.test_data_audio_features)})
        compression_opts = dict(method='zip',
                                archive_name='out.csv')
        df.to_csv('out.zip', index=False,
                  compression=compression_opts)

    def main(self, learning_curves=True, export_test_label=True):

        self.import_data()

        self.split_dataset()

        self.extract_features()

        # Tags Feature
        print("Tags Features")
        self.pre_process_tags()
        self.train_predict_evaluate(self.kbest_Tags, self.train_data_labels,
                                    self.kbest_validationTags, self.valid_data_labels, mnb=True)

        # Metadata Feature
        print("\nMetadata Features")
        self.train_predict_evaluate(self.train_data_metadata_features, self.train_data_labels,
                                    self.valid_data_metadata_features, self.valid_data_labels, mnb=False)

        # Title Feature
        print("\nTitle Features")
        self.pre_process_title()
        self.train_predict_evaluate(self.kbest_title, self.train_data_labels,
                                    self.kbest_validationTitle, self.valid_data_labels, mnb=True)

        # Audio Features
        print("\nAudio Features")
        self.train_predict_evaluate(self.train_data_audio_features, self.train_data_labels,
                                    self.valid_data_audio_features, self.valid_data_labels, mnb=False)

        if export_test_label:
            self.export_test_labels()

        if learning_curves:
            # Draw Learning Curves
            cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

            # Multi Layer Perceptron Learning Curves
            model = MLPClassifier(hidden_layer_sizes=(
                1000, 5), solver='sgd', random_state=6, learning_rate_init=0.01, learning_rate="constant", max_iter=1000)
            estimator = make_pipeline(StandardScaler(), model)
            self.draw_curves(estimator, self.train_data_audio_features, self.train_data_labels, ylim=(0.05, 1.5),
                             cv=cv, n_jobs=4)

            # Naive Bayes Learning Curves
            estimator = GaussianNB()
            self.draw_curves(estimator, self.train_data_audio_features, self.train_data_labels, ylim=(0.05, 1.5),
                             cv=cv, n_jobs=4)

            # Logistic Regression Learning Curves
            estimator = make_pipeline(
                StandardScaler(), LogisticRegression(max_iter=2000))
            self.draw_curves(estimator, self.train_data_audio_features, self.train_data_labels, ylim=(0.05, 1.5),
                             cv=cv, n_jobs=4)

            plt.show()


# Runs the code
machine = Music_Genre_Prediction()
machine.main(learning_curves=True, export_test_label=True)
```
